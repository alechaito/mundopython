```python
# initialize my_set
conjunto = {1, 3, 5, 7}
print(conjunto)

#adicionar um elemento ao conjunto
conjunto.add(2)
print(conjunto)

# adicionando multiplos elementos
conjunto.update([9, 11, 0])
print(conjunto)

#removendo e descartando elementos
conjunto.discard(11)
conjunto.remove(9)
print(conjunto)
```

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

#uniao dos conjuntos A e B
#operador pipe
print(A | B)

print(A.union(B))
print(B.union(A))
```

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# e empresarial
print(A & B)

print(A.intersection(B))
print(B.intersection(A))
```

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

#operador menos
print(A - B)

#somente o que pertence exclusivamente ao A
print(A.difference(B))
#somente o que pertence exclusivamente ao B
print(B.difference(A))
```

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

#operador potencia
print(A ^ B)

#o que nao está em A e B
print(A.symmetric_difference(B))
#o que nao está em A e B
print(B.symmetric_difference(A))
```
